```python
import numpy as np
from sklearn.naive_bayes import BernoulliNB

# 1. Prepare the Training Data from the blog's table
# Features: [Contains "free", Contains "buy", Contains "hurry"]
# 1 = Yes, 0 = No
X_train = np.array([
    [1, 0, 1],  # Email 1: Yes, No, Yes
    [0, 1, 0],  # Email 2: No, Yes, No
    [1, 1, 1],  # Email 3: Yes, Yes, Yes
    [0, 0, 0]   # Email 4: No, No, No
])

# Labels: 1 = Spam, 0 = Not Spam
y_train = np.array([1, 0, 1, 0])

# 2. Initialize the Bernoulli Naive Bayes Classifier
# We use BernoulliNB because our features are binary (0 or 1)
model = BernoulliNB()

# 3. Train the model
model.fit(X_train, y_train)

# 4. Predict for a New Email
# Example: An email contains "free" and "buy" but NOT "hurry" -> [1, 1, 0]
new_email = np.array([[1, 1, 0]])
prediction = model.predict(new_email)
probability = model.predict_proba(new_email)

# 5. Output the result
result = "Spam" if prediction[0] == 1 else "Not Spam"
print(f"Prediction for the new email: {result}")
print(f"Probability (Not Spam vs Spam): {probability[0]}")
```
